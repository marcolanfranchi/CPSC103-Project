We chose to represent each field in our compound in our data definitions:

 - Although the Date (MonthByYear) will not have an effect on the line that our graph will produce, we chose to include a data definition for it. 

 - The RiskFreeRate is the first variable that we need as a benchmark to represent the risk free interest rate at the given time period and will be used in calculating the expected return of a stock which will be one of the lines on our graph.

 - The ReturnOnMarket represents the return of the S&P 500 at the given date, which is a proxy for the overall market return.

 - The ReturnOnIbm is the historical return on IBM which will be used as a comparison tool for the expected return on IBM we calculate using the CAPM model. This will be represented in our graph as the historical return of the stock and will be comapred to the expected return of the stock.

 - Beta is another input to the CAPM formula to find the volatility of the IBM stock at the given time which will be represented by the expected return of a stock on our graph.

```python
from cs103 import *
from typing import NamedTuple, List
import csv


# MonthByYear

MonthByYear = str
#interp. the date by a month/year combination

MBY1 = '9/1/2020'
MBY2 = '4/1/2020'
MBY3 = '7/1/2017'

@typecheck
def fn_for_month_by_year (mby: MonthByYear) -> ...:
    return ... (mby) # template based on Atomic Non-Distinct


# RiskFreeRate

RiskFreeRate = float # in range [0, ...)
# interp. the risk free rate

RFR1 = 0
RFR2 = 0.35
RFR3 = 2.3

@typecheck
def fn_for_risk_free_rate(rfr: RiskFreeRate) -> ...:
    return ... (rfr) # template based on Atomic Non-Distinct


# ReturnOnMarket

ReturnOnMarket = float # in range [-100,100]
# interp. the return on the S&P 500 index, which is a proxy for the market

ROM1 = -10.30
ROM2 = 5.33
ROM3 = 16.40

@typecheck
def fn_for_return_on_market(rom: ReturnOnMarket) -> ...:
    return ... (rom) # template based on Atomic Non-Distinct


# ReturnOnIbm

ReturnOnIbm = float # in range [-100, 100]
# interp. the return on IBM's stock

ROI1 = -15.34
ROI2 = 6.04
ROI3 = -2.30

@typecheck
def fn_for_return_on_ibm(roi: ReturnOnIbm) -> ...:
    return ... (roi) #template based on Atomic Non-Distinct


# Beta

Beta = float # in range [0, ...)
# interp. the beta of the IBM stock, which is a measure of IBM stock's volatility relative
# to the market

B1 = 0.8
B2 = 1.2
B3 = 1.70

@typecheck
def fn_for_beta(b: Beta) -> ...:
    return ... (b) #template based on Atomic Non-Distinct



ReturnsByMonth = NamedTuple('ReturnsByMonth', [('month_by_year', MonthByYear),      
                                               ('risk_free_rate', RiskFreeRate),    
                                               ('return_on_market', ReturnOnMarket),   
                                               ('return_on_ibm', ReturnOnIbm),    
                                               ('beta', Beta)])             
# interp. the returns of a stock by month with the month and year, risk free rate,
# percentage return from the S&P 500, precentage return from the IBM, and the beta.

RBM1 = ReturnsByMonth('9/1/2020', 0.085, -3.92, -0.04, 1.172)
RBM2 = ReturnsByMonth('4/1/2020', 0.09, 12.68, 13.19, 1.264)
RBM3 = ReturnsByMonth('7/1/2017', 1.053, 1.93, -5.95, 0.9807)

@typecheck
def fn_for_returns_by_month(rbm: ReturnsByMonth) -> ...: # template based on compound 
    return ...(rbm.month_by_year,
               rbm.risk_free_rate,
               rbm.return_on_market,
               rbm.return_on_ibm,
               rbm.beta)
    

# List[ReturnsByMonth]
# interp. a list of Returns by Month

LORBM0 = []
LORBM1 = [RBM1, RBM2, RBM3]

@typecheck
def fn_for_lorbm(lorbm: List[ReturnsByMonth]) -> ...: # template based on arbitrary-sized
    # description of the acc                           # and the reference rule 
    acc = ... # type: ...
    for rbm in lorbm:
        acc = ...(acc, fn_for_returns_by_month(rbm))
    return ...(acc)

```

```python
###########
# Functions

import csv
from matplotlib import pyplot
import matplotlib.ticker as plticker
import matplotlib.font_manager


@typecheck
def main(fn: str) -> None:
    """
    Reads the file from given filename, plot 3 lines for: IBM historical return, 
    IBM return based on CAPM, difference between historical and CAPM return
    with a percentage return and date for the y and x-axis respectively and return None
    """
    # Template from HtDAP, based on composition
    return plot_line_chart(read(fn))


sfont = {'fontname':'serif'}

@typecheck
def plot_line_chart(lorbm: List[ReturnsByMonth]) -> None:
    """
    display a line chart of the historical stock returns, expected stock returns (calculated),
    and variance (calculated) in lorbm over time
    """
    #return None #stub
    # template based on visualization
    
    # limit x-axis ticks and enlarge chart (continued on line 49-50)
    fig, ax = pyplot.subplots(figsize=(25,14))
    
    # add grid on chart (left commented out for submission)
    pyplot.grid(True, linewidth=0.45, color='grey', linestyle='-')
    
    pyplot.xlabel('Date', fontsize = 23, **sfont)
    pyplot.ylabel('Return (in %)', fontsize = 23, **sfont)
    pyplot.title('IBM: Expected Return (CAPM) vs. Historical Return', fontsize = 30, **sfont)
    
    all_return_on_capm = get_return_on_capm(lorbm)
    all_historical_return = get_historical_return(lorbm)
    all_difference_in_return = get_difference_in_return(lorbm)
    time_in_months_by_year = get_months_by_year(lorbm)
    
    capm = ax.plot(time_in_months_by_year, all_return_on_capm)
    hist = ax.plot(time_in_months_by_year, all_historical_return)
    diff = ax.plot(time_in_months_by_year, all_difference_in_return)
    
    # displays every 14th value on x-axis ()
    loc = plticker.MultipleLocator(base=14.0) # this locator puts ticks at regular intervals
    ax.xaxis.set_major_locator(loc)
    
    #x and y tick font size
    pyplot.xticks(fontsize=16)
    pyplot.yticks(fontsize=16)
    
    #invert x-axis to set plots from oldest to newest
    pyplot.gca().invert_xaxis()
    
    pyplot.setp(capm, color='#000e42', linewidth=1.5, marker="s", label="Based on CAPM")
    pyplot.setp(hist, color='#00499b', linewidth=1.5, marker="s", label="Historical")
    pyplot.setp(diff, color='#af585e', linewidth=1.11, marker="|", label="Difference")
    
    # we want to show the legend because this plot contains two lines
    pyplot.legend(loc='upper left', prop={'size': 24})

    pyplot.show()
    
    return None


@typecheck
def read(filename: str) -> List[ReturnsByMonth]:
    """    
    reads information from the specified file and returns the Returns by Month
    """
    #return []  #stub
    # Template from HtDAP
    
    # lorbm contains the result so far
    lorbm = [] # type: List[ReturnsByMonth]
    with open(filename) as csvfile:
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            rbm = ReturnsByMonth(row[0], (parse_float(row[1])), (parse_float(row[2])), 
                                 (parse_float(row[3])), (parse_float(row[4])))
            lorbm.append(rbm)
    
    return lorbm


@typecheck
def get_months_by_year(lorbm: List[ReturnsByMonth]) -> List[str]:
    """
    return all the months by year from lorbm
    """
    #return [] #stub
    # template from List[ReturnsByMonth]
    # acc contains all of the returns by month seen so far
    acc = []    # type: List[str]
    for rbm in lorbm:
        acc.append(rbm.month_by_year)
    return acc


@typecheck
def get_return_on_capm(lorbm: List[ReturnsByMonth]) -> List[float]:
    """
    return all the returns based on the capm formula in percentage form (as displayed in get_capm_return(rbm))
    """
    #return [] #stub
    # template from List[ReturnsByMonth]
    # acc contains all of the returns by month seen so far
    acc = []    # type: List[float]
    for rbm in lorbm:
        acc = acc + [get_capm_return(rbm)]
    return acc


@typecheck
def get_capm_return(rbm: ReturnsByMonth) -> float:
    """
    returns the return based on the capm formula (risk-free rate + beta*(market return - risk-free rate)), 
    multiplied by 100 to be shown as a percentage
    """
    #return 0.00 #stub
    # template from ReturnsByMonth
    return (rbm.risk_free_rate + rbm.beta*(rbm.return_on_market - rbm.risk_free_rate))*100


@typecheck
def get_historical_return(lorbm: List[ReturnsByMonth]) -> List[float]:
    """
    return all the historical IBM returns from lorbm in percentage form (as displayed in historical_return(rbm))
    """
    #return [] #stub
    # template from List[ReturnsByMonth]
    # acc contains all of the returns by month seen so far
    acc = []    # type: List[float]
    for rbm in lorbm:
        acc = acc + [historical_return(rbm)]
    return acc


@typecheck
def historical_return(rbm: ReturnsByMonth) -> float:
    """
    return the historical IBM return from rbm, multiplied by 100 to be shown as a percentage
    """
    #return 0.00 #stub
    # template from ReturnsByMonth
    return (rbm.return_on_ibm)*100


@typecheck
def get_difference_in_return(lorbm: List[ReturnsByMonth]) -> List[float]:
    """
    return the absolute difference in percentages between the CAPM calculation for IBM's return 
    and IBM's historical return
    """
    #return [] #stub
    # template from List[ReturnsByMonth]
    # acc contains all of the returns by month seen so far
    acc = []    # type: List[float]
    for rbm in lorbm:
        acc = acc + [abs(get_capm_return(rbm) - historical_return(rbm))]
    return acc


start_testing()

# This should produce a graph with x-axis: "Date",
# y-axis "Return (in %)", a legend indicating the three lines
# displayed on the graph: Return based on CAPM, Historical return
# and the difference between the two, as well as the title
# "IBM: Expected Return (CAPM) vs. Historical Return".
#
# Here, we can see that for a time period before 9/1/2020, 
# the CAPM return was higher than the historical return.
expect(main('Stock_Return_Data_test1.csv'), None)

# This should produce a graph like the one above but with a combination of different
# lines. (A sketch of what that should look like is in the markdown cell below.)
expect(main('Stock_Return_Data_Test(2).csv'), None)

# This should produce a blank graph with x-axis: "Date",
# y-axis "Return (in %)", a legend indicating the three lines
# displayed on the graph: Return based on CAPM, Historical return
# and the difference between the two, as well as the title
# "IBM: Expected Return (CAPM) vs. Historical Return".
expect(plot_line_chart(LORBM0), None)

# This should produce a graph like the one above but with a series of
# points at 4/1/2020 and 7/1/2017
expect(plot_line_chart([RBM2, RBM3]), None)


# Examples and tests for read
expect(read('Stock_Return_Data_test1.csv'), [ReturnsByMonth('9/1/2020', 0.085, -3.92, -0.04, 1.172),
                                               ReturnsByMonth('8/1/2020', 0.095, 7.01, 0.30, 1.202),
                                               ReturnsByMonth('7/1/2020', 0.083, 5.51, 1.80, 1.217)])


expect(read('Stock_Return_Data_Test(2).csv'), [ReturnsByMonth('4/1/2018', 1.763, 0.27, -5.52, 0.9002),
                                               ReturnsByMonth('3/1/2018', 1.67, -2.69, -0.57, 0.9367),
                                               ReturnsByMonth('2/1/2018', 1.618, -3.89, -4.81, 0.945)])


# examples and tests for get_months_by_year
expect(get_months_by_year(LORBM0), [])
expect(get_months_by_year(LORBM1), ['9/1/2020', '4/1/2020', '7/1/2017'])

# examples and tests for get_return_on_capm
expect(get_return_on_capm(LORBM0), [])
expect(get_return_on_capm(LORBM1), [-460.88599999999997, 1600.376, 191.30739])
expect(get_return_on_capm([RBM1]), [-460.88599999999997])
expect(get_return_on_capm([RBM2, RBM3]), [1600.376, 191.30739])

# examples and tests for get_capm_return
expect(get_capm_return(RBM1), -460.88599999999997)
expect(get_capm_return(RBM2), 1600.376)
expect(get_capm_return(RBM3), 191.30739)

# examples and tests for get_historical_return
expect(get_historical_return(LORBM0), [])
expect(get_historical_return(LORBM1), [-4.0, 1319.0, -595.0])
expect(get_historical_return([RBM2, RBM3, RBM1]), [1319.0, -595.0, -4.0])
expect(get_historical_return([RBM2, RBM1]), [1319.0, -4.0])

# examples and tests for historical_return
expect(historical_return(RBM1), -4.00)
expect(historical_return(RBM2), 1319.00)
expect(historical_return(RBM3), -595.00)

# examples and tests for get_difference_in_return
expect(get_difference_in_return(LORBM0), [])
expect(get_difference_in_return(LORBM1), [456.88599999999997, 281.376, 786.3073899999999])


summary()
```
